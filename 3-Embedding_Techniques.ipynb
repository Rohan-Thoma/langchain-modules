Lets load the speech.txt document and get the embeddings for that:

1. First load the text document ( Data Ingestion )
2. Split that in to chunks using RecursiveCharacterTextSplitter ( Data Trasnformation )
3. Convert the chunks in to embeddings ( Data embedding in to vectors )
4. Store the vector embeddings to a vector_store_DB


```python
#Loading the document
from langchain_community.document_loaders import TextLoader

text_loader = TextLoader("Data/speech.txt")

text_doc = text_loader.load()

text_doc
```

```python
#Now split it in to chunks
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap= 50)

#Here we have already converted the text file into document format, so we will use split_documents()
final_documents = text_splitter.split_documents(text_doc)
final_documents
```

Now we will store these vectors in to the database

```python
# Vector embedding and vector storeDB
# Here we will use ChromaDB ( There are many database options which we will see later )
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(final_documents,embeddings_1000)
db
```

Now, we will retrive the relevant document from the database which is similar to the user query and perform a vector similarity search and then get the results

```python
query = " It will all be easier for us to conduct ourselves as belligerents"
retrieved_results = db.similarity_search(query)
print(retrieved_results)
```

##### Sentence Transformers on Hugging Face

Hugging Face sentence-transformers is a Python framework for state-of-the-art sentence, text and image embeddings. One of the embedding models is used in the HuggingFaceEmbeddings class. We have also added an alias for SentenceTransformerEmbeddings for users who are more familiar with directly using that package.

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name = "all-MiniLM-L6-v2"
)
```

```python
text = "This is a test document"
query_result = embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

Similarly we can also give a list of documents 

```python
#Here we will create a list of 2 sentences and then give it a different function called embed_documents
doc_result = embeddings.embed_documents([text, "This is test document 2"])
doc_result[0]
```
