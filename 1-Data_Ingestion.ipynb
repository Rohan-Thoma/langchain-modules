3. Web based Loader
* This will actually enable us to read websites ( like selenium and all )

```python
from langchain_community.document_loaders import WebBaseLoader

webpage_loader = WebBaseLoader(web_paths=("https://lilianweng.github.io/posts/2024-02-05-human-data-quality/",),)
#Here if we use web_paths , we can pass multiple of them as tuple, as we are passing only 1, we will put a comma
#Also there was an error saying set useragent ( like the browser agent etc ), we will just put comma and leave so that default browser agent will actually gets taken
```

```python
page_data = webpage_loader.load() # Here we can see that we are actually able to load the content of the given webpage
page_data
```

Now, lets see how to get the specific content of a webpage like I only want to get the title.
* For that we need to install beautiful soup library as well
* We can directly pass the parmeters of this library on to the langchain, webpage loader to get what we need

```python
from langchain_community.document_loaders import WebBaseLoader
import bs4 #beautiful soup

webpage_loader = WebBaseLoader(web_paths=("https://lilianweng.github.io/posts/2024-02-05-human-data-quality/",),
                               bs_kwargs=dict(parse_only=bs4.SoupStrainer(
                                   class_=("post-title","post-content","post-header")
                               )))
```

```python
page_data_new = webpage_loader.load()
page_data_new
```

4. Arxiv loader
* We can directly load the reasearch papers directly from their website by giving the document number

```python
from langchain_community.document_loaders import ArxivLoader

#Here we will directly get the docs by adding .load() at the end
paper_docs = ArxivLoader(query = "1605.08386", load_max_docs=2).load() # This is the citation number 
len(paper_docs)
```

```python
paper_docs
```
