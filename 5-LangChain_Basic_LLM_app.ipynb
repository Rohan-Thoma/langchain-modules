### 1. Setting up the API's required

```python
import os
from dotenv import load_dotenv,dotenv_values

load_dotenv() #We will load all the env variables
#Once we ran this, we do not need to run os.getenv() again, because we have already loaded the keys to the current process memory

values = dotenv_values(".env")  # returns a dict of key→value from that file, we check whether we have loaded the keys or not
print(values.keys()) #with this we can confirm, we have loaded the openai api key
```

Langchain API key is specifically used for Langsmith tracking , also we have an additional key called langchain tracing v2 which should be set to true. We will see why for all this below.


### 2. Models in langchain

```python
from langchain.chat_models import init_chat_model

model = init_chat_model(model = "gpt-4.1", api_key = values['OPENAI_API_KEY'])

response = model.invoke("what is generative AI ?")

response
```

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(api_key=values['OPENAI_API_KEY'], model = "gpt-4.1")
#Here we need to pass the api_key like this , but as we have already loaded the dotenv apikey with the same format ( no name changes )\
#It will be picked up automatically

print(llm)
```

```python
result = llm.invoke("What is generative AI ?")
```

### 3. Chat Prompt Template:

Here it is telling the LLM to behave in a certain manner.

```python
##chatpromt template
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are an expert AI engineer. Provide me answers based on the question"),
        ("user", "{input}")
    ]
)

prompt
```

```python
#chain
chain = prompt|llm

response = chain.invoke({"input":"Can you tell me about Langsmith?"})
print(response)
```

```python
type(response)
```

```python
response.content
```

### 4. String output parser

* This is like getting the output from the llm and how we want to display it. 
* There are many formats or types of it, but also we can create our own custom output parser

```python
#String output parser
from langchain_core.output_parsers import StrOutputParser

#This output parser will be responsible in displaying my entire output
output_parser = StrOutputParser()
chain = prompt|llm|output_parser


response = chain.invoke({"input": "Can you tell me about Lnagsmith?"})
print(response)
```
